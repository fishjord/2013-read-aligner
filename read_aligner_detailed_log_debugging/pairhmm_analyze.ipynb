```python
from collections import Counter
```

```python
edges = {}
vertices = {}

states = ["M_t", "Ir_t", "Ir_g", "M_u", "Ir_u", "Ir_g"]

edges['start_M_t'] = []
vertices['start_M_t'] = { 'cov' : 0, 'nodes' : []}

uniq_nodes = { 0 : (None, 'start', 0, 0, 0, 0)}

for line in open("./oom_details.txt"):
    lexemes = [x.split("=")[-1] for x in line.split(" ")]

    node_idx = int(lexemes[0], 16)
    prev = int(lexemes[1], 16)
    state = int(lexemes[2])
    kmer = lexemes[3]
    cov = int(lexemes[4])
    seqidx = int(lexemes[6])
    score = float(lexemes[7])
    fscore = float(lexemes[8])

    label = kmer + '_' + states[state]
    prev_label = uniq_nodes[prev][1] + '_' + states[uniq_nodes[prev][2]]
    if label not in edges:
        edges[label] = []
        vertices[label] = {'cov' : cov, 'nodes' : []}
    
    vertices[label]['nodes'].append(node_idx)
    uniq_nodes[node_idx] = (prev, kmer, state, score, fscore, seqidx)
    
    edges[prev_label].append(label)

#curr: 0x7fff19163eb0 0  state=0 CCGGATGCGGCGTTAACGCCTTATCCGGCC GGCCGGATAAGGCGTTAACGCCGCATCCGG cov=3 emission=1 seqidx=64 score=-2.65204 fscore=0 bits_saved=62.3022
```

```python
degree = {l : len(x) for l, x in edges.iteritems()}
print Counter(degree.values())
```

```python
from collections import Counter 
from IPython.display import Image

dot_stream = open("graph.dot", "w")
print >>dot_stream, "digraph {"
for vertex in sorted(edges.keys()):
    print >>dot_stream, '  {0} [label="{0} {1}"];'.format(vertex.replace('-', '_'), degree[vertex])
    print >>dot_stream
    for neighbor, count in Counter(edges[vertex]).iteritems():
        print >>dot_stream, '  {0} -> {1} [label="{2}"];'.format(vertex.replace('-', '_'), neighbor.replace('-', '_'), count)
        
    print >>dot_stream
    
print >>dot_stream, "}"
dot_stream.close()
```

```python
v = "TGCCGGATGCGGCGTTAACGCCTTATCCGG_Ir_t"

def path_len(node_idx, uniq_nodes):
    ret = 0
    while uniq_nodes[node_idx][0] != None:
        ret += 1
        node_idx = uniq_nodes[node_idx][0]
    return ret

path_sizes = {}
for node_idx in vertices[v]['nodes']:
    path_length = path_len(node_idx, uniq_nodes)
    if path_length not in path_sizes:
        path_sizes[path_length] = []
        
    path_sizes[path_length].append(node_idx)
    
print ["{0} {1}".format(k, len(v)) for k, v in path_sizes.iteritems()]
```

```python
def print_node(n, uniq_nodes):
    node = uniq_nodes[n]
    print "prev: {0} kmer: {1} state: {2} score: {3} fscore: {4} seqidx: {5}".format(node[0], node[1], states[node[2]], node[3], node[4], node[5])


n = path_sizes[27][0]
while n != None:
    print_node(n, uniq_nodes)
    n = uniq_nodes[n][0]
```
